# OOP Concept in Python

#### What is OOP

In procedural programming we code as sequence of steps. Procedural programming is great for data analysis and scripting.
In Object Oriented concept code as interactions of object. It is great for buidling frameworks and tools. It make the code more organizable and maintainable. The OOP concept is about objects and classes. Class is a blueprint on which a object is created. Classes has logical existence. Object is the instance of the class, it has physical existence. An object incorporate about state and behaviour. For example if a customer is a object then state(attribute) might be phone number and email id and behaviour(methods) might be place order or cancel order.

The state and behaviour bundled together in a class, the tendency is called **Encapsulation**. Classes defines the possible states and behaviours that the object of the class can have. In python everything is a object and every object has a class. The class can be founded by type method. State and behaviours are contained in attribute and methods respectively.

**Inheritance**

It is fundamentally important for code reuse. Class inheritance is about to inheriting a parent class and adding some modifications to it. 

            class Child(Parent):
                pass
                
Child class can use all attributes and methods from parent class. isinstance(child) should return Parent. Inheritance shows 'is a' relationship. We can add attributes, methods and constructor to a child class. We can run the constructor from the parent class by calling 'Parent.\_\_init\_\_()'. This should be defined inside child class init method. Also pass self to both the constructors. Then we can add more functionality. We can call the parent constructor, but we don't have to. We can use methods from child as well as parent class in the child class instances. Inside childs methods we can also call the parent class methods. 

**Operator Overloading**

Even if all attributes are same, then obj1 == obj2 will return False. Thats because of both the objects stored in different locations. We compare references while comparaing 2 objects. But thats not the case in numpy array or pandas dataframe.

However python allows us to overload all the comparing operator (ge, gt, lt, le, ne, eq). Python can implicitly call eq for ne, if ne is not available.

**Polymorphism**

Using a unified interface to operate on objects of different classes. This has to be true syntatically and semantically. 
Syntactically: function signatures are compatible like arguments and returns.
Semantically : The state of the object and program remains consistent. Like subclass methods doesn't strengthen input condition. Subclass method doesn't weaken output condition. And no output condition. 

**Managin data access**

All data in python are public by default. Because python philosophy is we all are adult here. We can restrict access by naming convention, using @property decorator or by modifying \_\_getattr\_\_() and \_\_setattr\_\_(). 

- In naming convention we use a single \_ to make the attribute privete. like \_attribute. It can be accessible publicly, but the developer try to mark this as a private attribute. As a class developer, use it for implementation details, helper function. 
- If an attribute start with \_\_ then this is private. Python does name mangling means the \_\_attr is treated same as \_attr. But a \_\_ says that it is not inherited. These are pseudo private attribute. It is used to prevent names in inherited classes. 
- Leading and trailing underscores are for inbuilt methods.

**Properties**

- Use a @property decorator on a method whose name is exactly the name of the restricted attribute, returns the restricted attribute.
- Use a @attr.setter on method attr that will be called on obj.attr = value. The value to assign is passed through argument. We can handle for the given attribute like checking for negative value and raising an error.
- Like for a employee class the salary is the restricted attribute. The salary method is with @property method return the salary. The method same salary method with @salary.setter will assign a new value to salary. 
- When we assign like obj.attr, then the setter method will be called by default.
- If you don't add a @attr.setter, the property will be read only like: df.shape
- Add @attr.getter when the property value is retrived.
- Add @attr.deleter when the property is deleted using del.
